```python
import pandas as pd
df = pd.read_csv('csv_files/customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#summary description using .describe()
df.describe(include='all')
```

```python
# Checking if missing data or null values are present in the datasetabs
df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#Checking if there is still any null statement in "Review Rating"
df.isnull().sum()
```

```python
#Now let's change the columns to lower case and also replace the spaces with underscore

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior' ]
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
#We have two colums that looks the same 'discount_applied', 'promo_code_used'
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
#Let's check if they are really all the same.
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#Since it is True, let's drop'promo_code_used' column 
df = df.drop('promo_code_used', axis=1)
```

```python
# Install required libraries

from sqlalchemy import create_engine
from urllib.parse import quote_plus

# SQL Server connection
username = "sa"
password = "your_password"
host = "localhost"
port = "1433"
database = "customer_behavior"

# Note: requires Microsoft ODBC Driver installed separately on your machine
driver = quote_plus("ODBC Driver 17 for SQL Server")
engine = create_engine(f"mssql+pyodbc://{username}:{password}@{host},{port}/{database}?driver={driver}")

# Write DataFrame to SQL Server
df.to_sql("customer", engine, if_exists="replace", index=False)

# Read back sample (SQL Server uses TOP instead of LIMIT)
pd.read_sql("SELECT TOP 5 * FROM customer;", engine)
```
